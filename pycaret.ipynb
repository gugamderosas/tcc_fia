```python
import numpy as np
import pandas as pd
import jinja2
from pycaret.classification import *
from sklearn.model_selection import train_test_split
```

```python
# Read Google Drive 
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Read abt

df = pd.read_csv('/content/drive/My Drive/TCC_FIA_IA/base_abt_modelagem.csv', index_col=0)
df
```

```python
# Rename variables

key_vars = ['ref_prev', 'Item']

num_vars = ['venfis', 'venfin', 'itens', 'itens_dist', 'datas', 'estoque', 'recencia']

target = 'revendeu_prox_mes'

cat_vars = []

features = num_vars+cat_vars
```

```python
X = df[features]
y = df[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    train_size=0.8, 
                                                    stratify= y, 
                                                    random_state=42)
```

```python
df_train = X_train
df_train['revendeu_prox_mes'] = y_train
df_train.head(2)
```

```python
df_test = X_test
df_test['revendeu_prox_mes'] = y_test
df_test.head(2)
```

```python
# Setup do modelo
exp_setup = setup(data = df_train , target = target, 
                   categorical_features = cat_vars, numeric_features = num_vars, session_id = 42)
```

```python
compare_models(fold = 5, sort = 'auc')
```

```python
opt = create_model('et', fold=5)
tuned_opt = tune_model(opt, fold=5, optimize='precision', n_iter = 30)
pred_opt = predict_model(tuned_opt, data = df_test)
```
